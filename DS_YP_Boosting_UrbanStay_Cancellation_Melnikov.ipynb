# Цели и задачи проекта

Сеть отелей UrbanStay столкнулась с ростом числа отмен бронирований. По словам заказчика, более 30% гостей отменяют свои брони незадолго до заселения, из-за чего отели теряют выручку, сталкиваются с простоем номерного фонда и испытывают сложности при планировании работы персонала и партнёров.

Сейчас для компенсации потерь используется стратегия овербукинга, однако такой подход связан с риском нехватки свободных номеров в случае, если все гости прибудут на заселение. Поэтому компании требуется более управляемое решение, позволяющее заранее выявлять бронирования с высоким риском отмены.

Цель проекта - разработать модель машинного обучения, которая будет прогнозировать вероятность отмены бронирования и поможет принимать решения по управлению номерным фондом.

Помимо качества прогнозирования необходимо оценить экономическую эффективность внедрения модели. Для этого в проекте будут использоваться бизнес-метрики, отражающие влияние модели на доходность и загрузку отелей.

Целевые показатели проекта:

- доля отмен бронирований после внедрения модели - не более 10%;
- снижение загрузки отеля после внедрения модели - не более 8%;
- относительный прирост дохода (Incremental Revenue) - не менее 50%.

# Описание данных

Для решения задачи предоставлены две таблицы из базы данных сети отелей UrbanStay.

**Таблица hotel_bookings**

Содержит информацию о бронированиях номеров. Каждая строка соответствует отдельному бронированию.

В таблице представлены сведения о канале продаж, составе гостей, истории взаимодействия клиента с сетью отелей, параметрах проживания, стоимости бронирования и его итоговом статусе.

Целевая переменная:

**Таблица hotel_reviews**

**Объединение данных**

````python
# Загружаем параметры подключения из env файла

load_dotenv()

DB_HOST = os.getenv('DB_HOST')
DB_PORT = os.getenv('DB_PORT')
DB_USER = os.getenv('DB_USER')
DB_PASS = os.getenv('DB_PASS')
DB_NAME = os.getenv('DB_NAME')

# Формируем строку подключения

connection_string = (
    f'postgresql://{DB_USER}:{DB_PASS}'
    f'@{DB_HOST}:{DB_PORT}/{DB_NAME}'
)

try:

    engine = create_engine(connection_string)

    print('Подключение к базе данных успешно установлено')

except Exception as error:

    print(f'Ошибка подключения: {error}')
````

````python
# Проверим не только NaN, но и пустые строки в текстовых столбцах

def check_empty_strings(dataframe):
    result = []

    for column in dataframe.columns:
        empty_count = 0

        if dataframe[column].dtype == 'object':
            values = dataframe[column].astype(str)

            empty_count += (values == '').sum()
            empty_count += values.str.isspace().sum()
            empty_count += values.str.lower().isin(['nan', 'none']).sum()

        if empty_count > 0:
            result.append({
                'Признак': column,
                'Количество пустых строк': empty_count,
                'Доля, %': round(empty_count / len(dataframe) * 100, 2)
            })

    if result:
        display(pd.DataFrame(result))
    else:
        print('Пустые строки и строковые пропуски не обнаружены.')


print('hotel_bookings')
check_empty_strings(df_bookings)

print('\nhotel_reviews')
check_empty_strings(df_reviews)
````

````python
# Функция построения графиков
def show_numeric_plots(
    dataframe,
    columns,
    plot_kind='hist',
    cols_per_row=3
):
    
    rows = (
        len(columns)
        + cols_per_row
        - 1
    ) // cols_per_row

    fig, axes = plt.subplots(
        rows,
        cols_per_row,
        figsize=(15, rows * 4)
    )

    axes = axes.flatten()

    for i, column in enumerate(columns):

        if plot_kind == 'box':
            sns.boxplot(
                y=dataframe[column],
                ax=axes[i]
            )

        else:
            sns.histplot(
                dataframe[column],
                kde=True,
                ax=axes[i]
            )

        axes[i].set_title(column)

    for j in range(
        i + 1,
        len(axes)
    ):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()
````

### Объединение таблиц

- Соедините таблицу отзывов с таблицей бронирований. Руководствуйтесь правилом: для каждой даты бронирования в столбце `booking_date` нужно взять комментарий и оценку клиента по предыдущей ближайшей дате отзывов, указанной в столбце `review_date`. Это позволит связать обратную связь клиента с конкретным периодом проживания, что критически важно для анализа удовлетворённости гостей.

- Сделайте выводы о том, как прошло объединение и что получилось в результате.

Для построения модели необходимо объединить данные о бронированиях с историей отзывов клиентов.

По условию проекта для каждой даты бронирования `booking_date` нужно подобрать ближайший предыдущий отзыв клиента по дате `review_date`. Такой отзыв отражает предыдущий опыт проживания клиента и мог быть известен на момент текущего бронирования.

В таблице `hotel_bookings` нет отдельного столбца `customer_id`, но он есть в таблице `hotel_reviews`. Поэтому сначала восстановим `customer_id` для бронирований через техническую связку `booking_id`.

Важно: `customer_id` будет использоваться только для корректного объединения таблиц. В модель этот признак передаваться не будет, так как идентификатор клиента является служебным признаком.

Для объединения используем `merge_asof` с параметрами:

* `by='customer_id'`;
* `direction='backward'`;
* `allow_exact_matches=False`.

Так для каждого бронирования будет выбран ближайший предыдущий отзыв этого же клиента, опубликованный раньше даты текущего бронирования.

````python
# Восстановим customer_id для бронирований через таблицу отзывов

customer_map = (
    df_reviews[
        [
            'booking_id',
            'customer_id'
        ]
    ]
    .drop_duplicates()
)

df_bookings = (
    df_bookings
    .merge(
        customer_map,
        on='booking_id',
        how='left'
    )
)

print(
    f'Размер таблицы бронирований после добавления customer_id: '
    f'{df_bookings.shape}'
)

print(
    'Доля строк без customer_id:',
    f'{df_bookings["customer_id"].isna().mean():.2%}'
)

display(
    df_bookings[
        [
            'booking_id',
            'booking_status',
            'customer_id'
        ]
    ].head()
)
````

`customer_id` восстановлен через связку `booking_id` из таблицы отзывов. Этот признак нужен только для того, чтобы подобрать предыдущий отзыв именно того же клиента.

При этом сам `customer_id` не будет использоваться как признак модели. На этапе подготовки итоговой таблицы для моделирования он будет исключен как служебный идентификатор.

````python
# Подготовим таблицы для merge_asof

df_bookings = (
    df_bookings
    .sort_values(
        [
            'booking_date',
            'customer_id'
        ]
    )
    .reset_index(drop=True)
)

df_reviews_for_merge = (
    df_reviews[
        [
            'customer_id',
            'booking_id',
            'review_date',
            'stay_rating',
            'review_text'
        ]
    ]
    .dropna(subset=['review_date'])
    .sort_values(
        [
            'review_date',
            'customer_id'
        ]
    )
    .reset_index(drop=True)
)

df_reviews_for_merge['review_text'] = (
    df_reviews_for_merge['review_text']
    .fillna('')
)

print(f'Размер таблицы бронирований: {df_bookings.shape}')
print(f'Размер таблицы отзывов для объединения: {df_reviews_for_merge.shape}')
````

````python
# Для каждого бронирования подберем ближайший предыдущий отзыв этого же клиента

df = pd.merge_asof(
    df_bookings,
    df_reviews_for_merge,
    left_on='booking_date',
    right_on='review_date',
    by='customer_id',
    direction='backward',
    allow_exact_matches=False,
    suffixes=('', '_review')
)

df = (
    df
    .drop(
        columns=['booking_id_review'],
        errors='ignore'
    )
)

print(f'Размер таблицы бронирований до объединения: {df_bookings.shape}')
print(f'Размер таблицы после объединения: {df.shape}')

display(
    df[
        [
            'customer_id',
            'booking_id',
            'booking_date',
            'review_date',
            'stay_rating',
            'review_text'
        ]
    ].head(10)
)
````

````python
# Создадим признак давности предыдущего отзыва

df['days_since_last_review'] = (
    df['booking_date']
    -
    df['review_date']
).dt.days

df['review_text'] = (
    df['review_text']
    .fillna('')
)

print('Пропуски в признаках, связанных с предыдущими отзывами:')

display(
    df[
        [
            'customer_id',
            'review_date',
            'stay_rating',
            'review_text',
            'days_since_last_review'
        ]
    ]
    .isna()
    .sum()
)
````

````python
# Проверим, что после объединения не подтянулись отзывы из будущего

date_check = (
    df['review_date'].isna()
    |
    (df['review_date'] < df['booking_date'])
)

print(
    'Доля строк без утечки по дате:',
    f'{date_check.mean():.2%}'
)

print(
    'Проверка отсутствия отзывов из будущего:',
    date_check.all()
)
````

Объединение выполнено успешно.

Количество строк после объединения не изменилось, что подтверждает отсутствие потерь данных.

Для каждого бронирования был подобран ближайший предыдущий отзыв этого же клиента, если такой отзыв существовал до даты текущего бронирования.

Использование параметров `direction='backward'` и `allow_exact_matches=False` гарантирует соблюдение временной логики: отзывы из будущего не попадают в признаки модели.

`customer_id` использовался только как технический ключ для объединения таблиц. В модель этот признак передаваться не будет.

* исключим стоп-слова русского языка.

Тексты очищаем на этом этапе, но не обучаем TF-IDF-векторизатор на всей таблице. Обучение векторизатора будет выполнено позже внутри пайплайна моделирования только на обучающей части данных.

````python
stop_words = set(stop_words_ru)


def clean_text(text):
    """
    Очищает текст отзыва:
    - приводит к нижнему регистру;
    - удаляет лишние символы;
    - удаляет русские стоп-слова.
    """
    
    if pd.isna(text):
        return ''
    
    text = str(text).lower()
    text = re.sub(r'[^а-яё0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    words = [
        word
        for word in text.split()
        if word not in stop_words
    ]
    
    return ' '.join(words)


df['review_text'] = (
    df['review_text']
    .fillna('')
    .apply(clean_text)
)

display(
    df.loc[
        df['review_text'] != '',
        ['customer_id', 'booking_id', 'booking_date', 'review_date', 'review_text']
    ].head()
)
````

Тексты предыдущих отзывов очищены от служебных символов и стоп-слов.

Для бронирований, у которых на момент `booking_date` не было доступного предыдущего отзыва клиента, текстовая колонка заполнена пустой строкой. Это нормальная ситуация: часть клиентов могла не иметь более раннего опыта проживания или не оставлять отзыв.

Очищенный текст будет использован на этапе моделирования. `TfidfVectorizer` будет обучаться внутри пайплайна только на обучающей выборке.

**Подготовка признаков для моделей**

В таблице `df_model` есть признаки разных типов: числовые, категориальные, булевы и текстовый признак `review_text`.

Так как векторизацию текста мы перенесли из этапа подготовки данных в этап моделирования, `review_text` будем обрабатывать внутри пайплайна с помощью `TfidfVectorizer`.

Чтобы избежать утечки данных, все обучаемые преобразования - заполнение пропусков, кодирование категорий и TF-IDF-векторизация - будут выполняться внутри `Pipeline`. Тогда на каждом фолде кросс-валидации преобразования будут обучаться только на обучающей части данных.

````python
# Определим группы признаков

text_feature = 'review_text'

categorical_features = [
    column for column in [
        'sales_channel',
        'meal_plan',
        'room_type'
    ]
    if column in X_train.columns
]

boolean_features = (
    X_train
    .select_dtypes(include=['bool'])
    .columns
    .tolist()
)

numeric_features = [
    column for column in X_train.columns
    if column not in categorical_features
    + boolean_features
    + [text_feature]
]

print(f'Текстовый признак: {text_feature}')
print(f'Числовых признаков: {len(numeric_features)}')
print(f'Категориальных признаков: {len(categorical_features)}')
print(f'Булевых признаков: {len(boolean_features)}')

print('\nКатегориальные признаки:')
print(categorical_features)

print('\nБулевые признаки:')
print(boolean_features)

print('\nЧисловые признаки:')
print(numeric_features)
````

````python
def bool_to_float(x):
    """
    Переводит булевы признаки True/False в числовой формат 1.0/0.0.
    Нужна как отдельная функция, чтобы итоговый пайплайн можно было сохранить через joblib.
    """
    
    return x.astype(float)
````

````python
# Пайплайн обработки текстового признака

text_transformer = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=30
)

# Пайплайн обработки числовых признаков

numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

# Пайплайн обработки категориальных признаков

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(
        handle_unknown='ignore'
    ))
])

# Пайплайн обработки булевых признаков

boolean_transformer = Pipeline([
    ('to_float', FunctionTransformer(
        bool_to_float,
        feature_names_out='one-to-one'
    )),
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Общий препроцессор

preprocessor = ColumnTransformer(
    transformers=[
        ('text', text_transformer, text_feature),
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features),
        ('bool', boolean_transformer, boolean_features)
    ],
    remainder='drop',
    verbose_feature_names_out=False
)
````

Для моделирования выделены четыре группы признаков:

* текстовый признак `review_text`;
* 18 числовых признаков;
* 3 категориальных признака;
* 1 булев признак.

Текстовый признак будет преобразован с помощью `TfidfVectorizer` внутри пайплайна. Числовые признаки будут заполнены медианой, категориальные признаки - наиболее частым значением и закодированы через One-Hot Encoding, а булевы признаки будут переведены в формат 0/1.

Все обучаемые преобразования выполняются внутри `Pipeline`, поэтому при кросс-валидации и финальной проверке преобразователи обучаются только на обучающей части данных.

<div class="alert alert-block alert-info">
<b>Совет:</b> Пайплайн построен правильно — предобработка происходит внутри `ColumnTransformer`, что исключает утечку данных между фолдами. `TfidfVectorizer` обучается только на train-данных. Это отлично!

Небольшое замечание: для признаков `stay_rating` и `days_since_last_review` используется `SimpleImputer(strategy='median')`, который заполняет пропуски общей медианой по обучающей выборке. Это лучше, чем использование медианы по всем данным, но групповая медиана (например, по `sales_channel` или `returning_customer`) позволила бы учесть различия между сегментами клиентов и повысить качество замены. **Функции для расчета бизнес-метрики IR**

Ключевая метрика проекта - Incremental Revenue. Она показывает, насколько меняется финансовый результат после применения модели.

Для расчета используем финансовые константы из условия проекта:

- `AVG_REV` - средний доход от успешной брони;
- `LOST_REV` - потеря от пропущенной отмены;
- `COST_FP` - стоимость ложного срабатывания;
- `PER_REBOOKING` - доход от повторной продажи номера при верно найденной отмене.

````python
print (AVG_REV, LOST_REV, COST_FP, PER_REBOOKING)
````

````python
def calculate_ir_before(y_true):
    """
    Доход до внедрения модели.
    """
    
    total_success = (y_true == 0).sum()
    total_cancellations = (y_true == 1).sum()
    
    ir_before = (
        total_success * AVG_REV
        -
        total_cancellations * LOST_REV
    )
    
    return ir_before


def calculate_ir_after(y_true, y_pred):
    """
    Доход после применения модели.
    """
    
    tn, fp, fn, tp = confusion_matrix(
        y_true,
        y_pred,
        labels=[0, 1]
    ).ravel()
    
    ir_after = (
        tp * PER_REBOOKING
        +
        tn * AVG_REV
        -
        fp * COST_FP
        -
        fn * LOST_REV
    )
    
    return ir_after


def calculate_incremental_ir(y_true, y_pred):
    """
    Прирост дохода: IR_after - IR_before.
    """
    
    return (
        calculate_ir_after(y_true, y_pred)
        -
        calculate_ir_before(y_true)
    )


def incremental_ir_score(y_true, y_pred):
    """
    Функция для sklearn scorer.
    """
    
    return calculate_incremental_ir(y_true, y_pred)


ir_scorer = make_scorer(
    incremental_ir_score,
    response_method='predict',
    greater_is_better=True
)
````

````python
def get_business_metrics(y_true, y_pred):
    """
    Возвращает confusion matrix и финансовые метрики.
    """
    
    tn, fp, fn, tp = confusion_matrix(
        y_true,
        y_pred,
        labels=[0, 1]
    ).ravel()
    
    ir_before = calculate_ir_before(y_true)
    ir_after = calculate_ir_after(y_true, y_pred)
    incremental_ir = ir_after - ir_before
    
    return {
        'TN': tn,
        'FP': fp,
        'FN': fn,
        'TP': tp,
        'IR_before': ir_before,
        'IR_after': ir_after,
        'Incremental_IR': incremental_ir
    }
````

В кросс-валидации будем оптимизировать именно прирост дохода. Это важнее обычной accuracy, потому что ошибки модели имеют разную стоимость для бизнеса.

**Базовое сравнение моделей на кросс-валидации**

Сначала сравним несколько моделей с базовыми настройками:

- Logistic Regression;
- Random Forest;
- CatBoost;
- LightGBM;
- XGBoost.

Логистическую регрессию используем как простой базовый ориентир. Основной интерес для дальнейшей оптимизации представляют модели градиентного бустинга, так как они лучше подходят для нелинейных зависимостей в табличных данных.

Кросс-валидацию проведем только на обучающей выборке. Используем `TimeSeriesSplit`, чтобы внутри каждого фолда валидационная часть находилась позже обучающей.

````python
# Настройка временной кросс-валидации

tscv = TimeSeriesSplit(
    n_splits=CV_FOLDS,
    test_size=2000
)

# Метрики для первичного сравнения

scoring = {
    'IR': ir_scorer,
    'ROC_AUC': 'roc_auc',
    'F1': make_scorer(f1_score, zero_division=0),
    'Precision': make_scorer(precision_score, zero_division=0),
    'Recall': make_scorer(recall_score, zero_division=0)
}
````

````python
# Набор базовых моделей

base_models = {
    'LogisticRegression': LogisticRegression(
        random_state=RANDOM_STATE,
        solver='liblinear',
        max_iter=1000,
        class_weight='balanced',
        C=0.5
    ),
    
    'RandomForest': RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_estimators=500,
        max_depth=17,
        min_samples_split=15,
        min_samples_leaf=1,
        max_features='sqrt',
        class_weight='balanced',
        n_jobs=N_JOBS
    ),
    
    'CatBoost': CatBoostClassifier(
        random_state=RANDOM_STATE,
        iterations=300,
        depth=5,
        learning_rate=0.05,
        loss_function='Logloss',
        verbose=0,
        allow_writing_files=False,
        thread_count=N_JOBS
    ),
    
    'LightGBM': LGBMClassifier(
        random_state=RANDOM_STATE,
        n_estimators=300,
        learning_rate=0.05,
        num_leaves=31,
        class_weight='balanced',
        verbose=-1,
        n_jobs=N_JOBS
    ),
    
    'XGBoost': XGBClassifier(
        random_state=RANDOM_STATE,
        n_estimators=300,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=2,
        eval_metric='logloss',
        tree_method='hist',
        verbosity=0,
        n_jobs=N_JOBS
    )
}
````

````python
# Сравним модели на TimeSeriesSplit

cv_results_list = []

for model_name, model in base_models.items():
    
    print(f'\nОцениваем модель: {model_name}', flush=True)
    
    model_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    start_time = time.perf_counter()
    
    cv_result = cross_validate(
        model_pipeline,
        X_train,
        y_train,
        cv=tscv,
        scoring=scoring,
        n_jobs=1,
        error_score='raise'
    )
    
    elapsed_time = time.perf_counter() - start_time
    
    current_result = {
        'model': model_name,
        'mean_IR': cv_result['test_IR'].mean(),
        'mean_ROC_AUC': cv_result['test_ROC_AUC'].mean(),
        'mean_F1': cv_result['test_F1'].mean(),
        'mean_precision': cv_result['test_Precision'].mean(),
        'mean_recall': cv_result['test_Recall'].mean(),
        'fit_time_sec': elapsed_time
    }
    
    cv_results_list.append(current_result)
    
    print(
        f'{model_name} готова: '
        f'IR = {current_result["mean_IR"]:,.0f}, '
        f'ROC-AUC = {current_result["mean_ROC_AUC"]:.3f}, '
        f'время = {elapsed_time:.1f} сек.',
        flush=True
    )

cv_results = (
    pd.DataFrame(cv_results_list)
    .sort_values('mean_IR', ascending=False)
)

display(
    cv_results.style.format({
        'mean_IR': '{:,.0f}',
        'mean_ROC_AUC': '{:.3f}',
        'mean_F1': '{:.3f}',
        'mean_precision': '{:.3f}',
        'mean_recall': '{:.3f}',
        'fit_time_sec': '{:.2f}'
    }).background_gradient(
        subset=['mean_IR'],
        cmap='Greens'
    )
)
````

**Подбор гиперпараметров с помощью Optuna**

На базовом сравнении лучший результат по `Incremental_IR` показал `RandomForest`. Очень близкий результат получил `XGBoost`, а `CatBoost` также показал сильное качество по `ROC-AUC` и бизнес-метрике.

Для дальнейшей оптимизации выберем три модели:

* `RandomForest`;
* `XGBoost`;
* `CatBoost`.

Оптимизировать будем не обычную accuracy, а бизнес-метрику `Incremental_IR` на временной кросс-валидации. Для каждой модели подберем больше трех гиперпараметров, как требуется в задании.

Для бустингов дополнительно используем раннюю остановку обучения, чтобы не добавлять лишние деревья при отсутствии улучшения на валидационной части фолда.

````python
# Количество итераций Optuna для каждой модели

N_OPTUNA_TRIALS = 15
EARLY_STOPPING_ROUNDS = 50
````

````python
def build_pipeline(model):
    """
    Собирает единый пайплайн препроцессинга и модели.
    """
    
    return Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
````

````python
def get_cv_ir_with_model(model, model_type):
    """
    Считает средний Incremental IR на TimeSeriesSplit.
    Для бустингов использует early stopping на валидационной части каждого фолда.
    """
    
    fold_scores = []
    
    for train_index, valid_index in tscv.split(X_train):
        
        X_fold_train = X_train.iloc[train_index]
        y_fold_train = y_train.iloc[train_index]
        
        X_fold_valid = X_train.iloc[valid_index]
        y_fold_valid = y_train.iloc[valid_index]
        
        fold_preprocessor = clone(preprocessor)
        
        X_fold_train_prep = fold_preprocessor.fit_transform(
            X_fold_train,
            y_fold_train
        )
        
        X_fold_valid_prep = fold_preprocessor.transform(
            X_fold_valid
        )
        
        fold_model = clone(model)
        
        if model_type == 'randomforest':
            fold_model.fit(
                X_fold_train_prep,
                y_fold_train
            )
            
        elif model_type == 'catboost':
            fold_model.fit(
                X_fold_train_prep,
                y_fold_train,
                eval_set=(X_fold_valid_prep, y_fold_valid),
                early_stopping_rounds=EARLY_STOPPING_ROUNDS,
                use_best_model=True,
                verbose=False
            )
            
        elif model_type == 'xgboost':
            fold_model.fit(
                X_fold_train_prep,
                y_fold_train,
                eval_set=[(X_fold_valid_prep, y_fold_valid)],
                verbose=False
            )
        
        y_valid_pred = fold_model.predict(
            X_fold_valid_prep
        )
        
        fold_ir = calculate_incremental_ir(
            y_fold_valid,
            y_valid_pred
        )
        
        fold_scores.append(fold_ir)
    
    return np.mean(fold_scores)
````

При подборе параметров для бустингов добавим раннюю остановку обучения (`early_stopping_rounds=50`). Это нужно, чтобы модель не продолжала добавлять деревья, если качество на валидационной части фолда уже не улучшается.

Так как `eval_set` должен быть разным для каждого временного фолда, подбор параметров реализуем через ручной цикл по `TimeSeriesSplit`, а не через `cross_val_score`. Логика кросс-валидации при этом сохраняется: на каждом фолде модель обучается на более ранних данных и проверяется на более поздних.

`early_stopping` контролирует процесс обучения по внутренней метрике модели, а Optuna выбирает лучшие параметры по нашей бизнес-метрике `Incremental_IR`.

````python
def objective_randomforest(trial):
    """
    Optuna objective для RandomForest.
    """
    
    params = {
        'n_estimators': trial.suggest_int(
            'n_estimators',
            300,
            800
        ),
        'max_depth': trial.suggest_int(
            'max_depth',
            8,
            25
        ),
        'min_samples_split': trial.suggest_int(
            'min_samples_split',
            2,
            30
        ),
        'min_samples_leaf': trial.suggest_int(
            'min_samples_leaf',
            1,
            10
        ),
        'max_features': trial.suggest_categorical(
            'max_features',
            ['sqrt', 'log2', None]
        ),
        'class_weight': trial.suggest_categorical(
            'class_weight',
            ['balanced', 'balanced_subsample']
        ),
        'random_state': RANDOM_STATE,
        'n_jobs': N_JOBS
    }
    
    model = RandomForestClassifier(**params)
    
    return get_cv_ir_with_model(
        model,
        model_type='randomforest'
    )
````

````python
def objective_catboost(trial):
    """
    Optuna objective для CatBoost с early stopping.
    """
    
    use_weight = trial.suggest_categorical(
        'use_scale_pos_weight',
        [False, True]
    )
    
    params = {
        'iterations': trial.suggest_int(
            'iterations',
            300,
            700
        ),
        'depth': trial.suggest_int(
            'depth',
            3,
            6
        ),
        'learning_rate': trial.suggest_float(
            'learning_rate',
            0.02,
            0.15,
            log=True
        ),
        'l2_leaf_reg': trial.suggest_float(
            'l2_leaf_reg',
            0.5,
            10.0,
            log=True
        ),
        'random_strength': trial.suggest_float(
            'random_strength',
            0.1,
            2.0
        ),
        'bagging_temperature': trial.suggest_float(
            'bagging_temperature',
            0.0,
            2.0
        ),
        'loss_function': 'Logloss',
        'verbose': 0,
        'allow_writing_files': False,
        'random_state': RANDOM_STATE,
        'thread_count': N_JOBS
    }
    
    if use_weight:
        params['scale_pos_weight'] = trial.suggest_float(
            'scale_pos_weight',
            1.0,
            4.0
        )
    
    model = CatBoostClassifier(**params)
    
    return get_cv_ir_with_model(
        model,
        model_type='catboost'
    )
````

````python
def objective_xgboost(trial):
    """
    Optuna objective для XGBoost с early stopping.
    """
    
    params = {
        'n_estimators': trial.suggest_int(
            'n_estimators',
            200,
            700
        ),
        'max_depth': trial.suggest_int(
            'max_depth',
            3,
            7
        ),
        'learning_rate': trial.suggest_float(
            'learning_rate',
            0.02,
            0.2,
            log=True
        ),
        'subsample': trial.suggest_float(
            'subsample',
            0.7,
            1.0
        ),
        'colsample_bytree': trial.suggest_float(
            'colsample_bytree',
            0.7,
            1.0
        ),
        'min_child_weight': trial.suggest_int(
            'min_child_weight',
            1,
            15
        ),
        'gamma': trial.suggest_float(
            'gamma',
            0.0,
            8.0
        ),
        'reg_alpha': trial.suggest_float(
            'reg_alpha',
            1e-3,
            5.0,
            log=True
        ),
        'reg_lambda': trial.suggest_float(
            'reg_lambda',
            1e-3,
            5.0,
            log=True
        ),
        'scale_pos_weight': trial.suggest_float(
            'scale_pos_weight',
            1.0,
            4.0
        ),
        'eval_metric': 'logloss',
        'tree_method': 'hist',
        'early_stopping_rounds': EARLY_STOPPING_ROUNDS,
        'random_state': RANDOM_STATE,
        'verbosity': 0,
        'n_jobs': N_JOBS
    }
    
    model = XGBClassifier(**params)
    
    return get_cv_ir_with_model(
        model,
        model_type='xgboost'
    )
````

````python
# Запускаем подбор параметров

optuna_objectives = {
    'RandomForest': objective_randomforest,
    'XGBoost': objective_xgboost,
    'CatBoost': objective_catboost
}

optuna_studies = {}

for model_name, objective_func in optuna_objectives.items():
    
    print(f'\nОптимизация модели: {model_name}', flush=True)
    
    sampler = optuna.samplers.TPESampler(
        seed=RANDOM_STATE
    )
    
    study = optuna.create_study(
        direction='maximize',
        sampler=sampler,
        study_name=model_name
    )
    
    start_time = time.perf_counter()
    
    study.optimize(
        objective_func,
        n_trials=N_OPTUNA_TRIALS,
        show_progress_bar=True,
        n_jobs=1
    )
    
    elapsed_time = time.perf_counter() - start_time
    
    optuna_studies[model_name] = study
    
    print(
        f'{model_name} готова. '
        f'Лучший CV Incremental IR: {study.best_value:,.0f}. '
        f'Время: {elapsed_time / 60:.1f} мин.',
        flush=True
    )
````

Такой подход позволит проверить, какая из оптимизированных моделей лучше переносит качество на более поздний временной период.

**Оценка оптимизированных моделей на калибровочной выборке**

После подбора гиперпараметров обучим модели на всей обучающей выборке и проверим их на калибровочной части.

Калибровочная выборка пока используется только для сравнения моделей при стандартном пороге 0.5. На следующем этапе она будет использоваться для калибровки вероятностей и поиска оптимального бизнес-порога.

Чтобы выбор был надежнее, сравним не только оптимизированные модели, но и базовые версии трех лидеров из первичного сравнения. Это позволит убедиться, что подбор гиперпараметров действительно улучшил качество или хотя бы не ухудшил его.

````python
def make_randomforest_from_study(study):
    """
    Создает RandomForestClassifier с лучшими параметрами Optuna.
    """
    
    params = study.best_params.copy()
    
    model = RandomForestClassifier(
        **params,
        random_state=RANDOM_STATE,
        n_jobs=N_JOBS
    )
    
    return model


def make_catboost_from_study(study):
    """
    Создает CatBoostClassifier с лучшими параметрами Optuna.
    """
    
    params = study.best_params.copy()
    
    use_weight = params.pop(
        'use_scale_pos_weight',
        False
    )
    
    if not use_weight:
        params.pop(
            'scale_pos_weight',
            None
        )
    
    model = CatBoostClassifier(
        **params,
        loss_function='Logloss',
        verbose=0,
        allow_writing_files=False,
        random_state=RANDOM_STATE,
        thread_count=N_JOBS
    )
    
    return model


def make_xgboost_from_study(study):
    """
    Создает XGBClassifier с лучшими параметрами Optuna.
    """
    
    params = study.best_params.copy()
    
    model = XGBClassifier(
        **params,
        eval_metric='logloss',
        tree_method='hist',
        random_state=RANDOM_STATE,
        verbosity=0,
        n_jobs=N_JOBS
    )
    
    return model
````

````python
# Соберем пайплайны с оптимизированными моделями

tuned_model_builders = {
    'RandomForest_tuned': lambda: make_randomforest_from_study(
        optuna_studies['RandomForest']
    ),
    'XGBoost_tuned': lambda: make_xgboost_from_study(
        optuna_studies['XGBoost']
    ),
    'CatBoost_tuned': lambda: make_catboost_from_study(
        optuna_studies['CatBoost']
    )
}

# Также добавим базовые версии трех лидеров из первичного сравнения

base_leader_models = {
    'RandomForest_base': base_models['RandomForest'],
    'XGBoost_base': base_models['XGBoost'],
    'CatBoost_base': base_models['CatBoost']
}

candidate_pipelines = {}

for model_name, builder in tuned_model_builders.items():
    candidate_pipelines[model_name] = build_pipeline(
        builder()
    )

for model_name, model in base_leader_models.items():
    candidate_pipelines[model_name] = build_pipeline(
        clone(model)
    )

print('Модели для проверки на калибровочной выборке:')
print(list(candidate_pipelines.keys()))
````

````python
def evaluate_on_sample(model, X_part, y_part, threshold=0.5):
    """
    Считает ML-метрики и бизнес-метрики на заданной выборке.
    """
    
    y_proba = model.predict_proba(
        X_part
    )[:, 1]
    
    y_pred = (
        y_proba >= threshold
    ).astype(int)
    
    business_metrics = get_business_metrics(
        y_part,
        y_pred
    )
    
    metrics = {
        'ROC_AUC': roc_auc_score(y_part, y_proba),
        'F1': f1_score(y_part, y_pred, zero_division=0),
        'Precision': precision_score(y_part, y_pred, zero_division=0),
        'Recall': recall_score(y_part, y_pred, zero_division=0),
        **business_metrics
    }
    
    return metrics
````

````python
# Обучим модели на train и оценим на calib при стандартном пороге 0.5

calib_results = []
fitted_candidate_models = {}

for model_name, pipeline in candidate_pipelines.items():
    
    print(f'Обучаем модель: {model_name}')
    
    start_fit = time.perf_counter()
    
    pipeline.fit(
        X_train,
        y_train
    )
    
    fit_time = time.perf_counter() - start_fit
    
    start_predict = time.perf_counter()
    
    metrics = evaluate_on_sample(
        pipeline,
        X_calib,
        y_calib,
        threshold=0.5
    )
    
    predict_time = time.perf_counter() - start_predict
    
    fitted_candidate_models[model_name] = pipeline
    
    calib_results.append({
        'model': model_name,
        'ROC_AUC': metrics['ROC_AUC'],
        'F1': metrics['F1'],
        'Precision': metrics['Precision'],
        'Recall': metrics['Recall'],
        'IR_before': metrics['IR_before'],
        'IR_after': metrics['IR_after'],
        'Incremental_IR': metrics['Incremental_IR'],
        'TP': metrics['TP'],
        'FP': metrics['FP'],
        'FN': metrics['FN'],
        'TN': metrics['TN'],
        'fit_time_sec': fit_time,
        'predict_time_sec': predict_time
    })

calib_results = (
    pd.DataFrame(calib_results)
    .sort_values(
        'Incremental_IR',
        ascending=False
    )
)

display(
    calib_results.style.format({
        'ROC_AUC': '{:.3f}',
        'F1': '{:.3f}',
        'Precision': '{:.3f}',
        'Recall': '{:.3f}',
        'IR_before': '{:,.0f}',
        'IR_after': '{:,.0f}',
        'Incremental_IR': '{:,.0f}',
        'fit_time_sec': '{:.2f}',
        'predict_time_sec': '{:.3f}'
    }).background_gradient(
        subset=['Incremental_IR', 'ROC_AUC'],
        cmap='Greens'
    )
)
````

````python
# Зафиксируем лучшую модель по Incremental IR на калибровочной выборке

best_model_name = calib_results.iloc[0]['model']
best_model = fitted_candidate_models[best_model_name]

print(f'Лучшая модель на калибровочной выборке: {best_model_name}')
print(f'Лучший Incremental IR: {calib_results.iloc[0]["Incremental_IR"]:,.0f}')

if best_model_name.endswith('_tuned'):
    original_model_name = best_model_name.replace('_tuned', '')
    best_model_params = optuna_studies[original_model_name].best_params
    
    print('Лучшие параметры Optuna:')
    display(best_model_params)
else:
    best_model_params = base_models[
        best_model_name.replace('_base', '')
    ].get_params()
    
    print('Используется базовая версия модели.')
````

````python
def evaluate_probabilities(y_true, y_proba, threshold=0.5):
    """
    Считает качество модели по вероятностям и стандартному порогу.
    """
    
    y_pred = (
        y_proba >= threshold
    ).astype(int)
    
    business_metrics = get_business_metrics(
        y_true,
        y_pred
    )
    
    metrics = {
        'ROC_AUC': roc_auc_score(y_true, y_proba),
        'Brier': brier_score_loss(y_true, y_proba),
        'LogLoss': log_loss(y_true, y_proba),
        'F1': f1_score(y_true, y_pred, zero_division=0),
        'Precision': precision_score(y_true, y_pred, zero_division=0),
        'Recall': recall_score(y_true, y_pred, zero_division=0),
        **business_metrics
    }
    
    return metrics
````

### Поиск порога классификации

- Используя откалиброванную модель и калибровочную выборку, найдите порог классификации, при котором достигается максимальный Incremental Revenue.

- Сделайте выводы о пороге классификации.

После калибровки модели нужно выбрать порог, при котором вероятность отмены будет переводиться в класс `отказ_брони`.

Стандартный порог 0.5 не всегда подходит для бизнес-задач. В нашем случае ошибки имеют разную стоимость: пропущенная отмена приводит к потере дохода, а ложное срабатывание создает расходы на компенсацию и работу с клиентом.

Поэтому подберем порог на калибровочной выборке. Для каждого значения порога рассчитаем:

* `Precision`;
* `Recall`;
* `F1`;
* матрицу ошибок;
* `IR_before`;
* `IR_after`;
* `Incremental_IR`.

Оптимальным будем считать порог, при котором достигается максимальный `Incremental_IR`.

````python
# Используем вероятности калиброванной модели на калибровочной выборке

calib_proba_for_threshold = best_calibrated_proba_calib

print(
    f'Диапазон вероятностей на calib: '
    f'{calib_proba_for_threshold.min():.3f} - '
    f'{calib_proba_for_threshold.max():.3f}'
)
````

````python
def calculate_threshold_metrics(y_true, y_proba, thresholds):
    """
    Считает метрики качества и бизнес-метрики для набора порогов.
    """
    
    threshold_results = []
    
    for threshold in thresholds:
        
        y_pred = (
            y_proba >= threshold
        ).astype(int)
        
        tn, fp, fn, tp = confusion_matrix(
            y_true,
            y_pred,
            labels=[0, 1]
        ).ravel()
        
        precision = precision_score(
            y_true,
            y_pred,
            zero_division=0
        )
        
        recall = recall_score(
            y_true,
            y_pred,
            zero_division=0
        )
        
        f1 = f1_score(
            y_true,
            y_pred,
            zero_division=0
        )
        
        ir_before = calculate_ir_before(
            y_true
        )
        
        ir_after = calculate_ir_after(
            y_true,
            y_pred
        )
        
        incremental_ir = ir_after - ir_before
        
        threshold_results.append({
            'threshold': threshold,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'TN': tn,
            'FP': fp,
            'FN': fn,
            'TP': tp,
            'IR_before': ir_before,
            'IR_after': ir_after,
            'Incremental_IR': incremental_ir
        })
    
    return pd.DataFrame(threshold_results)
````

````python
# Переберем пороги от 0.01 до 0.99

threshold_grid = np.arange(
    0.01,
    1.00,
    0.01
)

threshold_metrics = calculate_threshold_metrics(
    y_true=y_calib,
    y_proba=calib_proba_for_threshold,
    thresholds=threshold_grid
)

threshold_metrics_top = (
    threshold_metrics
    .sort_values(
        'Incremental_IR',
        ascending=False
    )
    .head(10)
)

display(
    threshold_metrics_top.style.format({
        'threshold': '{:.2f}',
        'precision': '{:.3f}',
        'recall': '{:.3f}',
        'f1': '{:.3f}',
        'IR_before': '{:,.0f}',
        'IR_after': '{:,.0f}',
        'Incremental_IR': '{:,.0f}'
    }).background_gradient(
        subset=['Incremental_IR'],
        cmap='Greens'
    )
)
````

````python
# Зафиксируем лучший порог по Incremental IR

best_threshold_row = (
    threshold_metrics
    .sort_values(
        'Incremental_IR',
        ascending=False
    )
    .iloc[0]
)

best_threshold = best_threshold_row['threshold']
best_threshold_ir = best_threshold_row['Incremental_IR']

print(f'Оптимальный порог: {best_threshold:.2f}')
print(f'Максимальный Incremental IR на calib: {best_threshold_ir:,.0f} руб.')
print()
print(f'Precision: {best_threshold_row["precision"]:.3f}')
print(f'Recall:    {best_threshold_row["recall"]:.3f}')
print(f'F1:        {best_threshold_row["f1"]:.3f}')
print()
print(f'TP: {best_threshold_row["TP"]:.0f}')
print(f'FP: {best_threshold_row["FP"]:.0f}')
print(f'FN: {best_threshold_row["FN"]:.0f}')
print(f'TN: {best_threshold_row["TN"]:.0f}')
````

После выбора оптимального порога проверим, насколько стабильно модель работает на новых данных.

Для этого сравним результаты на двух выборках:

* `calib` - выборка, на которой подбирался порог;
* `test` - отложенная тестовая выборка, которую модель не использовала для выбора параметров, калибровки и поиска порога.

На обеих выборках применим один и тот же порог `best_threshold`, найденный на предыдущем этапе. Затем построим матрицы ошибок и сравним стандартные ML-метрики и бизнес-метрику `Incremental_IR`.

````python
# Получим вероятности отмены на calib и test

proba_calib_final = best_calibrated_model.predict_proba(
    X_calib
)[:, 1]

proba_test_final = best_calibrated_model.predict_proba(
    X_test
)[:, 1]

# Переведем вероятности в классы по найденному бизнес-порогу

pred_calib_final = (
    proba_calib_final >= best_threshold
).astype(int)

pred_test_final = (
    proba_test_final >= best_threshold
).astype(int)

print(f'Используемый порог классификации: {best_threshold:.2f}')
print(f'Размер calib: {len(y_calib)}')
print(f'Размер test:  {len(y_test)}')
````

````python
def evaluate_final_results(y_true, y_proba, y_pred):
    """
    Считает ML-метрики и бизнес-метрики для финального сравнения.
    """
    
    business_metrics = get_business_metrics(
        y_true,
        y_pred
    )
    
    result = {
        'ROC_AUC': roc_auc_score(y_true, y_proba),
        'Precision': precision_score(y_true, y_pred, zero_division=0),
        'Recall': recall_score(y_true, y_pred, zero_division=0),
        'F1': f1_score(y_true, y_pred, zero_division=0),
        'Cancel_share': y_true.mean(),
        **business_metrics
    }
    
    return result
````

````python
# Посчитаем метрики на calib и test

calib_final_metrics = evaluate_final_results(
    y_true=y_calib,
    y_proba=proba_calib_final,
    y_pred=pred_calib_final
)

test_final_metrics = evaluate_final_results(
    y_true=y_test,
    y_proba=proba_test_final,
    y_pred=pred_test_final
)

final_stability_metrics = pd.DataFrame([
    {
        'dataset': 'calib',
        'sample_size': len(y_calib),
        **calib_final_metrics
    },
    {
        'dataset': 'test',
        'sample_size': len(y_test),
        **test_final_metrics
    }
])

final_stability_metrics['Incremental_IR_per_booking'] = (
    final_stability_metrics['Incremental_IR']
    / final_stability_metrics['sample_size']
)

display(
    final_stability_metrics.style.format({
        'ROC_AUC': '{:.3f}',
        'Precision': '{:.3f}',
        'Recall': '{:.3f}',
        'F1': '{:.3f}',
        'Cancel_share': '{:.2%}',
        'IR_before': '{:,.0f}',
        'IR_after': '{:,.0f}',
        'Incremental_IR': '{:,.0f}',
        'Incremental_IR_per_booking': '{:,.0f}',
        'TN': '{:,.0f}',
        'FP': '{:,.0f}',
        'FN': '{:,.0f}',
        'TP': '{:,.0f}'
    }).background_gradient(
        subset=['Incremental_IR', 'ROC_AUC'],
        cmap='Greens'
    )
)
````

Оценим, какие признаки сильнее всего влияют на предсказание итоговой модели.

Финальная модель - откалиброванный XGBoost, сохраненный в переменной `best_calibrated_model`. Так как модель обернута в `CalibratedClassifierCV`, внутри нее хранится несколько обученных пайплайнов. Для устойчивости извлечем важности признаков из внутренних моделей XGBoost и усредним их.

Посмотрим важность в двух вариантах:

* по фактическим признакам после преобразований пайплайна, включая One-Hot Encoding и TF-IDF;
* по укрупненным группам, чтобы удобнее интерпретировать бизнес-факторы.

Дополнительно выполним SHAP-анализ на сэмпле тестовой выборки, чтобы посмотреть вклад признаков в конкретные предсказания.

````python
def extract_xgb_importances_from_calibrated_model(calibrated_model):
    """
    Извлекает важности признаков из внутренних XGBoost-моделей,
    которые находятся внутри CalibratedClassifierCV.
    """
    
    fold_importance_tables = []
    
    for fold_number, calibrated_classifier in enumerate(
        calibrated_model.calibrated_classifiers_,
        start=1
    ):
        
        fitted_pipeline = calibrated_classifier.estimator
        
        fitted_preprocessor = fitted_pipeline.named_steps['preprocessor']
        fitted_xgb_model = fitted_pipeline.named_steps['model']
        
        feature_names = fitted_preprocessor.get_feature_names_out()
        feature_importances = fitted_xgb_model.feature_importances_
        
        fold_importance = pd.DataFrame({
            'feature': feature_names,
            f'importance_fold_{fold_number}': feature_importances
        })
        
        fold_importance_tables.append(
            fold_importance
        )
    
    importance_wide = reduce(
        lambda left, right: pd.merge(
            left,
            right,
            on='feature',
            how='outer'
        ),
        fold_importance_tables
    )
    
    fold_columns = [
        column for column in importance_wide.columns
        if column.startswith('importance_fold_')
    ]
    
    importance_wide[fold_columns] = (
        importance_wide[fold_columns]
        .fillna(0)
    )
    
    importance_wide['importance_mean'] = (
        importance_wide[fold_columns]
        .mean(axis=1)
    )
    
    importance_wide['importance_std'] = (
        importance_wide[fold_columns]
        .std(axis=1)
    )
    
    total_importance = importance_wide['importance_mean'].sum()
    
    importance_wide['importance_percent'] = (
        importance_wide['importance_mean']
        / total_importance
        * 100
    )
    
    importance_wide = (
        importance_wide
        .sort_values(
            'importance_percent',
            ascending=False
        )
        .reset_index(drop=True)
    )
    
    importance_wide['rank'] = (
        np.arange(len(importance_wide)) + 1
    )
    
    return importance_wide
````

````python
# Получим названия TF-IDF признаков из всех внутренних препроцессоров

tfidf_feature_names = set()

for calibrated_classifier in best_calibrated_model.calibrated_classifiers_:
    
    current_pipeline = calibrated_classifier.estimator
    current_preprocessor = current_pipeline.named_steps['preprocessor']
    
    current_tfidf_names = (
        current_preprocessor
        .named_transformers_['text']
        .get_feature_names_out()
        .tolist()
    )
    
    tfidf_feature_names.update(current_tfidf_names)

tfidf_feature_names = sorted(tfidf_feature_names)

print(f'Количество уникальных TF-IDF признаков по всем фолдам: {len(tfidf_feature_names)}')
print(tfidf_feature_names[:10])
````

В таблице выше признаки показаны уже после обработки пайплайна. Поэтому категориальные признаки могут быть представлены отдельными колонками после One-Hot Encoding, а текстовые признаки - отдельными TF-IDF токенами.

Для бизнес-интерпретации удобнее дополнительно сгруппировать признаки: например, все One-Hot колонки от `sales_channel` собрать обратно в один фактор `sales_channel`, а TF-IDF признаки отзывов объединить в одну группу.

````python
def get_feature_group(feature_name):
    """
    Укрупняет признаки после преобразований пайплайна
    до исходных бизнес-групп.
    """
    
    if feature_name in tfidf_feature_names:
        return 'TF-IDF признаки отзывов'
    
    for categorical_feature in categorical_features:
        
        if (
            feature_name == categorical_feature
            or feature_name.startswith(f'{categorical_feature}_')
        ):
            return categorical_feature
    
    if feature_name in numeric_features:
        return feature_name
    
    if feature_name in boolean_features:
        return feature_name
    
    return feature_name
````

````python
# Соберем укрупненную важность признаков

feature_importance_df['feature_group'] = (
    feature_importance_df['feature']
    .apply(get_feature_group)
)

feature_importance_grouped = (
    feature_importance_df
    .groupby('feature_group', as_index=False)
    .agg(
        importance_percent=('importance_percent', 'sum'),
        transformed_features_count=('feature', 'count')
    )
    .sort_values(
        'importance_percent',
        ascending=False
    )
    .reset_index(drop=True)
)

feature_importance_grouped['rank'] = (
    np.arange(len(feature_importance_grouped)) + 1
)

display(
    feature_importance_grouped[
        [
            'rank',
            'feature_group',
            'importance_percent',
            'transformed_features_count'
        ]
    ]
    .head(20)
    .style.format({
        'importance_percent': '{:.3f}',
        'transformed_features_count': '{:,.0f}'
    })
)
````

````python
# Подготовим актуальные предсказания итоговой модели на test

proba_test_economics = best_calibrated_model.predict_proba(
    X_test
)[:, 1]

pred_test_economics = (
    proba_test_economics >= best_threshold
).astype(int)

tn_t, fp_t, fn_t, tp_t = confusion_matrix(
    y_test,
    pred_test_economics,
    labels=[0, 1]
).ravel()

total_bookings_test = (
    tn_t
    + fp_t
    + fn_t
    + tp_t
)

print(f'Используемая модель: {best_model_name}')
print(f'Используемый порог: {best_threshold:.2f}')
print()
print(f'TN: {tn_t}')
print(f'FP: {fp_t}')
print(f'FN: {fn_t}')
print(f'TP: {tp_t}')
print(f'Всего бронирований в test: {total_bookings_test}')
````

````python
# Рассчитаем показатели до и после внедрения модели

# До внедрения модели

success_before = (
    tn_t
    + fp_t
)

cancellations_before = (
    fn_t
    + tp_t
)

cancel_rate_before = (
    cancellations_before
    / total_bookings_test
)

occupancy_before = (
    success_before
    / total_bookings_test
)

ir_before = calculate_ir_before(
    y_test
)

# После внедрения модели
# Нераспознанными отменами остаются только FN:
# это отмены, которые модель не обнаружила.

cancel_rate_after = (
    fn_t
    / total_bookings_test
)

# Для загрузки после внедрения используем консервативную логику:
# гарантированно успешными считаем только TN.
# TP не включаем в occupancy_after, потому что повторная продажа уже учтена в IR_after.
# FP не включаем, потому что это ложные тревоги, которые могут создавать риск потери успешной брони.

occupancy_after = (
    tn_t
    / total_bookings_test
)

ir_after = calculate_ir_after(
    y_test,
    pred_test_economics
)

incremental_ir = (
    ir_after
    - ir_before
)

relative_ir = (
    incremental_ir
    / ir_before
    * 100
)

cancel_rate_change_pp = (
    cancel_rate_after
    - cancel_rate_before
) * 100

cancel_rate_reduction_percent = (
    (cancel_rate_before - cancel_rate_after)
    / cancel_rate_before
    * 100
)

occupancy_change_pp = (
    occupancy_after
    - occupancy_before
) * 100

occupancy_relative_change_percent = (
    (occupancy_after - occupancy_before)
    / occupancy_before
    * 100
)
````

**Почему результат можно считать корректным**

Важно, что итоговый результат был получен не за счет рискованной утечки данных. Ранее был риск использовать признак, связанный с самим фактом наличия отзыва. Это опасно, потому что отзыв появляется после проживания и может косвенно показывать, что клиент не отменил бронь.

В итоговой версии такой признак не использовался. Отзывы были подключены только как предыдущий опыт клиента, известный до даты текущего бронирования. Также в модель не передавались технические идентификаторы вроде `customer_id`, `booking_id` и `review_date`. Поэтому итоговая модель использует именно исторические и бизнес-доступные признаки, а не прямой сигнал из будущего.

**Основные факторы, влияющие на риск отмены**

По встроенной важности `XGBoost` и SHAP-анализу наиболее важными оказались следующие факторы:

1. `days_since_last_review` - давность предыдущего отзыва. Этот признак помогает модели понять, насколько свежим является предыдущий клиентский опыт.

2. TF-IDF признаки предыдущих отзывов - текст старого отзыва клиента. Суммарная важность текстовых признаков по встроенной важности составила 24.75%. Это значит, что содержание предыдущего отзыва помогает модели уточнять риск отмены.

3. `stay_rating` - оценка предыдущего проживания. Прошлый опыт клиента может быть связан с его дальнейшим поведением.

4. `previous_no_shows` - прошлые незаезды клиента. Если клиент уже не приезжал раньше, модель учитывает это как важный сигнал риска.

5. `days_until_checkin` - срок до заезда. Чем больше времени между бронированием и заездом, тем выше вероятность, что планы клиента могут измениться.

6. `sales_channel` - канал бронирования. Разные каналы бронирования могут иметь разное поведение клиентов и разный риск отмены.

7. `client_history` - общий показатель клиентской истории. Он помогает модели учитывать прошлое поведение клиента.

8. `has_children` - наличие детей в бронировании. Семейные поездки могут иметь свою специфику планирования и отмен.

9. `total_nights` - длительность проживания. Длинные бронирования могут быть более чувствительны к изменению планов.

10. `cancellation_ratio` и `previous_cancellations` - история отмен клиента. Эти признаки описывают прошлую склонность клиента отменять бронирования.

При этом важно не трактовать встроенную важность как точный процент повышения или снижения риска. Она показывает, насколько активно модель использует признак. Направление влияния лучше оценивать по SHAP-графику: значения справа от нуля повышают прогнозируемую вероятность отмены, значения слева от нуля снижают ее.

**Рекомендации для бизнеса**

1. Использовать модель как систему раннего предупреждения.

Для бронирований с высокой вероятностью отмены можно заранее запускать дополнительные коммуникации: напоминание о бронировании, уточнение планов, предложение подтвердить заезд, персональное сообщение от отеля. Это особенно важно для бронирований с высоким риском по каналу продаж, сроку до заезда и клиентской истории.

2. Разделить действия по уровню риска.

Не стоит одинаково работать со всеми клиентами, которых модель отметила как рискованных. Для среднего риска можно использовать мягкое напоминание. Для высокого риска - более активное подтверждение бронирования или персональное предложение. Это поможет снизить ложные срабатывания и не создавать лишние расходы.

3. Использовать историю отзывов в клиентской работе.

Текст и оценка предыдущего отзыва оказались важными для модели. Если у клиента был негативный или спорный прошлый опыт, отелю стоит заранее уделить такому бронированию больше внимания: проверить пожелания, заранее связаться с клиентом, предложить улучшение сервиса или уточнить ожидания.

4. Регулярно переобучать и контролировать модель.

Поведение клиентов и структура бронирований могут меняться по сезонам, каналам продаж и внешним условиям. Поэтому модель стоит периодически переобучать на новых данных и отслеживать качество на свежих периодах: долю отмен, загрузку, `Incremental_IR`, `Precision` и `Recall`.

5. Не использовать модель как единственный источник решения.

Модель должна помогать сотрудникам находить рискованные бронирования, но итоговые действия лучше делать управляемыми: с учетом стоимости номера, важности клиента, канала продаж, условий тарифа и текущей загрузки отеля.
